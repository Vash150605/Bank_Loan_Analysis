```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('raw_loan_data.csv')
print(df.info()) 
```

```python
median_income = df['Income'].median()
df['Income'] = df['Income'].fillna(median_income)
```

```python
df['Employment_Length_Years'] = df['Employment_Length_Years'].replace(-1, 0)
```

```python
# Check if data is clean
print(df.isnull().sum())
```

```python
# Plot 1: Compare Default vs Paid
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Loan_Status')
plt.title('Loan Status Breakdown (0=Paid, 1=Default)')
plt.show()
```

```python
# Plot 2: Average Interest Rate by Loan Status
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Loan_Status', y='Interest_Rate')
plt.title('Interest Rate vs Loan Default')
plt.show()
```

```python
df.to_csv('cleaned_loan_data.csv', index=False)
```
